# Evens

```python
# evens.py
#
# Takes a list of integers and returns a new list with only the even 
# numbers from the original list, in the same relative order.
# Usage:
#      % python evens.py
#
# Anastasia Erofeeva
# 10/03/19


from typing import List


def evens(lst: List[int]) -> List[int]:
    "Return the even elements, in order"
    
    # Empty list to store results
    res = []
    
    # Check each number in the given list
    for number in lst:
        
        # If the number is even, add it to the new list
        if number % 2 == 0:
            res.append(number)
    
    return res
```

## Unit Tests for evens

```python
lst = []
result = evens(lst)
expected = []
assert (result == expected), f'{lst} - Expected {expected}; got {result}'

lst = [1, 3, 5, 7]
result = evens(lst)
expected = []
assert (result == expected), f'{lst} - Expected {expected}; got {result}'

lst = [2, 3, 1, 4, 1, 2, 4, 5, 6, 9, 10]
result = evens(lst)
expected = [2, 4, 2, 4, 6, 10]
assert (result == expected), f'{lst} - Expected {expected}; got {result}'

print('Passed')
```

# Inorder

```python
# inorder.py
#
# Takes a list and returns whether or not it is in order.
# Usage:
#      % python inorder.py
#
# Anastasia Erofeeva
# 10/03/19


from typing import List


def inorder(lst: List) -> bool:
    """Is this list in order?"""
    
    try:
        
        # Check each index in the list
        for i in range(len(lst) - 1):

            # If element at this index is greater than next element, return False
            if lst[i] > lst[i + 1]:
                return False

        # Otherwise, the list is in order, so return True
        return True

    # If the items in the list are not the same type
    except TypeError:
        return('List items must be the same type')
```

## Unit Tests for inorder()

```python
assert(inorder([1, 4, 9, 13]) == True)
assert(inorder([1]) == True)
assert(inorder([]) == True)
assert(inorder([3, 1, 4]) == False)
assert(inorder([3, 2, 1]) == False)
assert(inorder(['one', 'ten', 'three', 'two']) == True)

assert(inorder([1, 4, 9, 13, 12]) == False)
assert(inorder(['one', 'two', 'three', 'four']) == False)

print('Pass')
```

# Unique

## Write a function unique that takes a list that may 
## have identical adjacent elements, and returns a 
## new list with the duplicates removed. 
### *Review the first Unit Test to see which duplicates to remove*

```python
# unique.py
#
# Takes a list and returns a new list without any adjacent duplicates.
# Usage:
#      % python unique.py
#
# Anastasia Erofeeva
# 10/03/19


from typing import List


def unique(lst: List) -> List:
    """Trim duplicates from list"""

    # If the list is empty, just return the list
    if len(lst) < 1:
        return lst
    
    # New empty list to store results
    res = []
    
    # Append first element from original list to the new list
    res.append(lst[0])
    
    # Check each index in the list, except index 0
    for i in range(1, len(lst)):
        
        # If element at this index differs from element at previous index
        if lst[i] != lst[i - 1]:
            
            # Add the element to the new list
            res.append(lst[i])
    
    return res
```

# Unit Tests

```python
lst = [1, 1, 2, 2, 1, 3, 3, 3, 3, 2, 1, 1]
result = unique(lst)
expected = [1, 2, 1, 3, 2, 1]
assert (result == expected), f'{lst} - Expected {expected}; got {result}'
    
lst = [3]
result = unique(lst)
expected = [3]
assert (result == expected), f'{lst} - Expected {expected}; got {result}'
    
lst = ['one', 'one', 'ten', 'three', 'two', 'two']
result = unique(lst)
expected = ['one', 'ten', 'three', 'two']
assert (result == expected), f'{lst} - Expected {expected}; got {result}'
    
lst = []
result = unique(lst)
expected = []
assert (result == expected), f'{lst} - Expected {expected}; got {result}'

print('Pass')
```

# Sum of Two

## Write a function to decides if we can write target as sum of two values

```python
# sum_of_two.py
#
# Takes an integer and a list of integers, and decides if the given integer
# is the sum of two different integers in the list.
# Usage:
#      % python sum_of_two.py
#
# Anastasia Erofeeva
# 10/04/19
#
# I got the idea to use index variables and a while loop to solve 
# this problem from: https://www.geeksforgeeks.org/given-an-array
# -a-and-a-number-x-check-for-pair-in-a-with-sum-as-x/


from typing import List


def sum_of_two(k: int, lst: List[int]) -> bool:
    """Is k the sum of two elements in the list?"""
    
    # Sort the list and save it as a new list
    sorted_list = lst.sort()
    
    first = 0                # Index of first element in list
    last = len(lst) - 1      # Index of last element in list
    
    # Check sum of elements in list, starting from first and last elements, 
    # then moving inward, if necessary. Repeat while first index is 
    # less than last index
    while first < last:
        
        # If sum of elements at first index and last index is k, return True
        if (lst[first] + lst[last] == k):
            return True
        
        # Else if sum of elements at first index and last index is
        # greater than k, move the last index lower by one
        elif (lst[first] + lst[last] > k):
            last -= 1
            
        # Else if sum of elements at first index and last index is
        # less than k, move the first index higher by one
        else:
            first += 1
            
    # No two numbers in the list sum to k, so return False
    return False
```

## Unit Tests for sum of two

```python
assert(sum_of_two(0, []) == False)
assert(sum_of_two(3, [3]) == False) 
assert(sum_of_two(3, [1, 2]) == True) 
assert(sum_of_two(17, [10, 15, 3, 7]) == True)
assert(sum_of_two(17, [10, 15, 4, 8]) == False) 

assert(sum_of_two(4, [2, 2]) == True)
assert(sum_of_two(4, [1, 2]) == False) 
assert(sum_of_two(4, [0, 4]) == True)
assert(sum_of_two(17, [1, 15, 3, 4, 5, 6, 7, 2]) == True)

print('Pass')
```

# Find Reversals
## Take a list, and find all pairs of reversals
## *Palindromes don't count*

```python
# reversals.py
#
# Takes a list of strings and returns a new list with only the strings
# whose reverse are also in the list.
# Usage:
#      % python reversals.py
#
# Anastasia Erofeeva
# 10/05/19
#
# I got the idea to use a list comprehension to make all the words 
# in the list lower case from: https://stackoverflow.com/questions/
# 42128170/how-to-lower-case-a-list-in-python?noredirect=1&lq=1
# I also looked up the syntax of the index() method here:
# https://www.programiz.com/python-programming/methods/list/index


from typing import List


def find_reversals(words: List[str]) -> List[str]:
    """Look for reversals in a list"""
    
    # Empty list to store results
    res = []
    
    # Make every word in the list lower case
    words = [word.lower() for word in words]
    
    # Check every word in the list
    for word in words:
        
        # Variable to store the reverse of the word
        reverse = word[::-1]
        
        # If reverse is in the original list
        if (reverse in words and 
            
            # And word is not already in the result list
            word not in res and 
            
            # And reverse is not in the result list
            reverse not in res and 
            
            # And reverse is not the same as word (not a palindrome)
            reverse != word):
            
            # Add word to the end of result list
            res.append(word)
            
    return res
```

## Unit Tests for find_reversals()

```python
assert(find_reversals(['art', 'Rat', 'Radar', 'scam', 'tar', 'vista']) == ['rat'])
assert(find_reversals(['art', 'Rat', 'Radars', 'scam', 'tartars', 'vista']) == [])

assert(find_reversals(['art', 'Rat', 'Radar', 'rat', 'radar', 'tar', 'tar']) == ['rat'])
assert(find_reversals(['Radar']) == [])
assert(find_reversals(['test']) == [])
assert(find_reversals([]) == [])

print('Pass')
```

```python
# reversed_words.py
#
# Takes a file of words, appends the words in the file to a list,
# and returns a new list with only the strings whose reverse are also 
# in the original list. 
# Usage:
#      % python reversed_words.py
#
# Anastasia Erofeeva
# 10/05/19
#
# I looked up the syntax of the sorted() method here:
# https://www.programiz.com/python-programming/methods/built-in/sorted


from typing import List


def find_reversals(words: List[str]) -> List[str]:
    """Look for reversals in a list"""
    
    # Empty list to store results
    res = []
    
    # Empty dictionary to store words and the reverse of the words
    d = dict()
    
    # Check each word in list
    for word in words:
        
        # Variable to store the reverse of the word
        reverse = word[::-1]
        
        # if word not in dictionary and word is not the same as reverse
        if (word not in d and word != reverse):
            
            # Add reverse-word key-value pair to dictionary
            d[reverse] = word
            
        # If word is one of dictionary keys
        if (word in d):
            
            # Append reverse of word to list of results
            res.append(reverse)  

    # Return sorted list of results
    return sorted(res)


def read_file(filename: str) -> List[str]:
    """Read the lines in a file and append to list"""
    
    # Open the file
    with open(filename, 'r') as f:
        
        # A list of the lines in the file, lower case, without \n characters
        return [word.lower().strip() for word in f.readlines()]


lst = read_file('words.txt')
revs = find_reversals(lst)
print(len(revs))

for word in revs:
    print(word)
```
